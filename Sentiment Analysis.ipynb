```python
from __future__ import division, unicode_literals

from math import ceil
from datetime import datetime, timedelta

import pandas as pd

from django.db import models
from pokeradio.scoring.models import Point
from django.contrib.auth.models import User
```

```python
WINDOW_LENGTH = 13
FRACTION = 0.2
```

```python
#today = datetime.today().date()
today = datetime(2014, 8, 8).date()
```

```python
def get_rs(user, today):
#    user = point.point_from
    window = [today - timedelta(days=WINDOW_LENGTH), today]
    
    this_week = Point.objects.filter(vote_from=user,created__range=window)
    votes_this_week = this_week.count()
    
    week_sum = this_week.aggregate(models.Sum('value'))['value__sum']
    week_sum = week_sum if week_sum else 0

    average = ceil(week_sum / 5)

    net_today = Point.objects.filter(vote_from=user,
                                     created__startswith=today)\
                                 .aggregate(models.Sum('value'))['value__sum']
    
    net_today = net_today if net_today else 0
    
    # Scale of difference required
    personal_threshold = ceil(votes_this_week / 5 * FRACTION) + 2

    difference_from_average = (net_today - average)

    happy = difference_from_average > (personal_threshold -1)
    sad = difference_from_average < -1 * personal_threshold
    
    if happy:
        face = ":)"
    elif sad:
        face = ":("
    else:
        face = ''
        
    # TODO Maybe sad should only be awarded on downvotes, and happy on upvotes
    # in case the user simply has not yet met their estimated target voting score for the day

    return user, average, net_today, "+- {0}".format(personal_threshold), difference_from_average, face, today
```

```python
data = []

today = datetime(2014, 8, 8).date()
for i in xrange(0, 10):
    today = today + timedelta(days=1)

    for u in User.objects.all():
        d = get_rs(u, today)

        if len(d[-1]) and d[2]:
            data.append(d)

df = pd.DataFrame(data = data, columns=['Name', 'Expected Today (av)', 'Actual today', 'Personal Threshold', 'Difference', 'Face', 'Date'])
```
